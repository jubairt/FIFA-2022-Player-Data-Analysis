### 11. Contract column

```python
fifa['Contract'].dtype
```

```python
fifa['Contract'].unique()
```

```python
for index, row in fifa.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
def extract_contract_info(contract):
    if contract == "Free" or "On Loan" in contract:
        start_date = np.nan
        end_date = np.nan
        contract_length = np.nan
    
    else:
        start_date, end_date = contract.split(' ~ ')
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_length = end_year - start_year
    
    return start_date, end_date, contract_length

# Applying function to contract column and creating new columns
new_cols = ['Contract Start','Contract End','Contract Length (Years)']
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i, column = new_cols[i], value=new_data[i])
```

```python
fifa[['Contract','Contract Start','Contract End','Contract Length (Years)']].isnull().sum()
```

```python
fifa[['Contract','Contract Start','Contract End','Contract Length (Years)']].dtypes
```

```python
fifa['Contract Start'] = pd.to_datetime(fifa['Contract Start'],errors='coerce')
fifa['Contract End'] = pd.to_datetime(fifa['Contract End'],errors='coerce')
```

```python
fifa[['Contract','Contract Start','Contract End','Contract Length (Years)']].dtypes
```

```python
fifa.sample(3)
```

```python
# Contract Categories

def categorize_contract_status(contract):
    if contract == "Free":
        return "Free"
    elif "On Loan" in contract:
        return "On Loan"
    else:
        return "Contract"
    
# Add contract status column
fifa.insert(fifa.columns.get_loc('Contract Length (Years)')+1, column="Contract Status", value=fifa['Contract'].apply(categorize_contract_status))
    
```

### 13. Height

```python
fifa['Height'].unique()
```

```python
def convert_height(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        
    return round(total_inches * 2.54)

# Applying into height column
fifa['Height'] = fifa['Height'].apply(convert_height)
```

```python
fifa = fifa.rename(columns = {'Height': 'Height(cm)'})
```

```python
fifa.sample(3)
```

### 14. Weight

```python
fifa['Weight'].dtype
```

```python
fifa['Weight'].unique()
```

```python
def convert_weight(weight):
    if "kg" in weight:
        return int(weight.strip("kg"))
    else:
        pounds = int(weight.strip("lbs"))
        return round(pounds/2.205)
    
# Applying function to weight column
fifa['Weight'] = fifa['Weight'].apply(convert_weight)
```

```python
fifa = fifa.rename(columns={'Weight':'Weight(kg)'})
```

### 77. Hits

Number of times the player has been searched for in the FIFA database.

```python
fifa['Hits'].dtype
```

```python
fifa['Hits'].unique()
```

```python
# there are missing values in this column, so we have to convert it into zero
fifa['Hits'] = fifa['Hits'].fillna(0)
fifa['Hits'] = fifa['Hits'].replace({'nan': 0})
fifa.isnull().sum()
```

```python
fifa['Hits'].unique()
```

```python
is_int_float = fifa[fifa['Hits'].apply(lambda x: isinstance(x, (float,int)))]
is_int_float['Hits'].count()
```

```python
def convert_Hits(hits):
    if isinstance(hits, str) and "K" in hits:
        return int(float(hits.strip("K").strip()) * 1000)
    else:
        return int(hits)
    
fifa['Hits'] = fifa['Hits'].apply(convert_Hits)
```
